```python
from projeto1 import global_grafo
import pandas as pd
import numpy as np

# 1. Remove a linha/coluna 0 (inicialização) e detecta o tamanho real
grafo_ajustado = np.array(global_grafo)[1:, 1:]  
n_vertices = grafo_ajustado.shape[0]

# 2. Define configurações visuais baseadas no tamanho
if n_vertices <= 30:
    cell_width = '25px'
    font_size = '9pt'
elif n_vertices <= 100:
    cell_width = '15px'
    font_size = '8pt'
else:  # +100 vértices
    cell_width = '10px'
    font_size = '7pt'

# 3. Cria o DataFrame com rótulos de 1 até n_vertices
df = pd.DataFrame(
    grafo_ajustado,
    index=range(1, n_vertices + 1),
    columns=range(1, n_vertices + 1)
)

# 4. Funções de formatação
def format_value(x):
    if pd.isna(x):
        return "-"
    elif np.isinf(x):
        return "∞"
    elif isinstance(x, (int, np.integer)):
        return str(x)
    elif isinstance(x, (float, np.floating)):
        return f"{x:.2f}" if not x.is_integer() else str(int(x))
    return str(x)

def background_style(x):
    return 'background-color: #e6f3ff' if x != 0 and not np.isinf(x) else 'background-color: white'

# 5. Estilo dinâmico (com barra de rolagem para +100 vértices)
styled_table = (
    df.style
    .map(background_style)
    .format(format_value)
    .set_properties(**{
        'text-align': 'center',
        'font-size': font_size,
        'padding': '2px',
        'border': '1px solid #ddd',
        'width': cell_width,
        'height': cell_width  # Mantém células quadradas
    })
    .set_table_styles([{
        'selector': 'table',
        'props': [
            ('width', 'auto'),
            ('margin', '0 auto'),
            ('display', 'block' if n_vertices > 100 else 'inline-block'),  # Barra de rolagem se necessário
            ('max-width', '100%'),
            ('overflow-x', 'auto' if n_vertices > 100 else 'visible')  # Barra horizontal para +100
        ]
    }, {
        'selector': 'th, td',
        'props': [('max-width', cell_width)]
    }])
)

styled_table
```
