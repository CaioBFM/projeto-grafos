```python
import pandas as pd

df = pd.read_csv("estatisticas.csv")

# Formatação dos valores
def format_value(x):
    return f"{int(x)}" if x == int(x) else f"{x}"

styled_df = (
    df.style
    .format({"VALOR": format_value})
    .set_caption("ESTÁTISTICAS BÁSICAS DO GRAFO:")
    .hide(axis="index")
)
```

```python
from IPython.display import display
display(styled_df)
```
